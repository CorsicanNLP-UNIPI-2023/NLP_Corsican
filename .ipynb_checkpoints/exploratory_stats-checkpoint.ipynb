```python
import pandas as pd
import nltk
import matplotlib.pyplot as plt
import glob
from bs4 import BeautifulSoup as bs
import re
```

```python
wiki = []
piazetta = []
bible = []

for f in glob.glob('./corpora/wikipedia_articles/*.xml'):
    wiki.append(f)

for f in glob.glob('./corpora/piazzetta_articles/*.xml'):
    piazetta.append(f)
    
for f in glob.glob('./corpora/corsican_bible/*.xml'):
    bible.append(f)
```

```python
def corpus_creation(file):
    with open(file, 'r') as f:
        content = f.readlines()
    
    content = "".join(content)
    soup = bs(content, "xml")
    
    la = soup.find_all('p')
    
    te = []
    for lala in la:
        te.append(lala.text)
    
    te = [re.sub(r'\n    ', '', t) for t in te]
    te = [re.sub(r'\n   ', '', t) for t in te]
    te = '. '.join(te)
    te = re.sub('\.\.', '.', te)
    
    return te
```

```python
wiki_corpus = []

for f in wiki:
    text = corpus_creation(f)
    wiki_corpus.append(text)
```

```python
test = ' '.join(wiki_corpus)
```

```python
tok = nltk.word_tokenize(test)
```

```python
len(tok)
```

```python
freq = nltk.FreqDist(tok)
```

```python
freq.plot(50, title="MFW 50")
```
